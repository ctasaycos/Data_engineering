```python
import duckdb
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
%load_ext sql
conn=duckdb.connect()
%sql conn
```

```python
%config SqlMagic.autopandas = True
%config SqlMagic.feedback = False
%config SqlMagic.displaycon = False
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ctasaycos/Data_engineering/main/Data/owid-covid-data.csv")
```

```python
query = """
select distinct index  from df
"""
df=duckdb.query(query).fetchdf()
```

```python
%%sql df_index <<
select 
cast(date as date) as date,
sum(new_deaths) new_deaths
from df
where cast(date as date)>'2019-01-01'
group by 1
order by date asc
```

```python
fig=px.line(df_index,x='date',y='new_deaths',title='Line chart',template='plotly_dark')
fig.show()
```

```python
%%sql complex_graph <<
select
continent,
location,
cast(sum(new_deaths) as int) as new_deaths
from df
where new_deaths between 1000 and 20000 and continent is not null
group by 1,2
```

```python
fig = px.bar(complex_graph, x='new_deaths', y='continent', color='location',
                 title='New Deaths by Continent and Location',
                 labels={'new_deaths': 'New Deaths', 'continent': 'Continent'},
                 template='plotly_dark')
fig.show()
```
